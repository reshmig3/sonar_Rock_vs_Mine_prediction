```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
sdf = pd.read_csv('/content/sonar data.csv',header=None)
sdf.shape
sdf.head()
```

```python
sdf.tail()
sdf.info()
```

```python
x=sdf[60].value_counts()
x=sdf.drop(columns=60,axis=1)
y=sdf[60]
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
x_train_prediction=model.predict(x_train)
training_a=accuracy_score(x_train_prediction,y_train)
print("training_data " ,training_a)
```

```python
x_test_prediction=model.predict(x_test)
test_a=accuracy_score(x_test_prediction,y_test)
print(test_a)
```
